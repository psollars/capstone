## Evaluation with BLEU

```python
with open(f"./cache/{retrieval_strategy}_{model}_responses.pickle", "rb") as handle:
    rag_responses = pickle.load(handle)
```

```python
for i in range(len(rag_responses["question"])):
    print(rag_responses["question"][i])
    print(">>>", rag_responses["ground_truth"][i])
    print("-" * 100)
    print(rag_responses["answer"][i])
    print("")
    print("=" * 100)
    print("=" * 100)
    print("")
```

```python
responses_df = pd.DataFrame(rag_responses)

responses_df["eye_check"] = [
    100, # 'Which class involves time series analysis?',
    100, # 'Who teaches the SQL and Databases class?',
    0, # 'What are the prerequisites for Data Science for Social Good?',
    0, # 'When are the office hours for the Math Methods course?',
    100, # 'Are there any weekly readings for Milestone II?',
    100, # 'What are the outcomes of Qualitative Inquiry?',
    100, # 'What textbook is required for SIADS 505?',
    100, # 'What textbook is required for Data Manipulation?',
    100, # 'Which week of unsupervised learning covers DBSCAN?',
    100, # 'How many credits are required to complete the MADS program?',
    100, # 'How long do students have to complete the MADS program start to finish?',
    100, # 'How many points is the comprehensive oral exam worth in SIADS 593?',
    100, # 'What is the penalty for late submission in SIADS 630?',
    100, # 'How do I get accommodations for a class?',
    100, # 'What is a backpack?',
    100, # 'When is the latest I can drop a course?',
    100, # 'How do I get an override to take a class?',
    100, # 'How do I take a leave of absence from the MADS program?',
    100, # 'What are the prerequisites for Search and Recommender Systems?'
]
```

```python
scored_responses = responses_df.copy()
scored_responses["bleu_score"] = scored_responses.apply(
    lambda x: sacrebleu.corpus_bleu([x["answer"]], [[x["ground_truth"]]]).score, axis=1
)

scored_responses.head()
```

```python
sns.heatmap(
    scored_responses[["bleu_score", "eye_check"]],
    yticklabels=scored_responses["question"].str[:42],
    annot=True,
    fmt=".4g",
    cmap=sns.cubehelix_palette(as_cmap=True),
)

plt.suptitle("BLEU Evaluation")
plt.title(f"Strategy: {retrieval_strategy} | Model: {model}", fontsize=10)

plt.show()
```
